This notebook will show how PCA works without the sklearn Class

```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv("./datasets/pca_data.csv")
```

```python
df.head()
```

```python
#mean centering

mean1= df['x1'].mean()
mean2= df['x2'].mean()
mean3= df['x3'].mean()


sd1= df['x1'].std()
sd2= df['x2'].std()
sd3= df['x3'].std()
```

```python
mean_centered_values= {
    '0': np.array([(x-mean1)/sd1 for x in df['x1']]),
    '1': np.array([(x-mean2)/sd2 for x in df['x2']]),
    '2': np.array([(x-mean3)/sd3 for x in df['x3']]),
}
```

```python
for i in range(3):
    df.iloc[:, i]= mean_centered_values[f'{i}']
```

```python
df= df.round(2)
```

```python
df
```

```python
x= df.iloc[:, :-1]
x.head()

```

```python
x.cov()
```

```python
cov_matrix = x.cov().values
cov_matrix.round(2
                 )
```

```python
# find the eigen vector and eigen value of cov matrix
eigen_value, eigen_vector= np.linalg.eig(cov_matrix)
```

```python
eigen_vector
```

```python
eigen_value
```

    - finally conveting the above 3 dim array into 2d. For 2d, we need two eigen vector and for 1d we need 1 eigen vector


```python
x.shape
```

```python
transformers= np.array([eigen_vector[0],eigen_vector[1]])
```

```python
transformers
```

```python
reduced_dim= np.dot(x.values, transformers.T)
```
