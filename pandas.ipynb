# Filtering the dataframe

```python
#for only one condition
mask= df['city']== 'Hyderabad'
mask
```

```python
#getting the infomation of having the city hyderabad
df[mask].head()
```

```python
#to check for more than one conditions, we need to create more than one mask
#eg to get the dataframe where city is Hyderabad and date < 2010-01-01

#mask to check the city
mask1= df["city"]== "Hyderabad"
mask2= df['date']< '2010-01-01'
```

```python
df[mask & mask2]  #use the bitwise operator because of boolean values
```

```python
df[mask2 | mask1]
```

# value_counts() function
    - This is generally used on categorical data or categorical values
    - only works on series and also returns a pandas series
    - Used to count the repetation of value in a column

```python
df['winner'].value_counts()
```

# Pandas plot function
    - here we must import another library to see it otherwise we cannot see it
    - There must be atleast one numerical column for it to work
    - You do not need to define x and y when plotting the result of .value_counts()

```python
import matplotlib.pyplot as plt
```

```python
df['toss_decision'].value_counts().plot(kind= 'bar', xlabel= "decision", ylabel= "number", title= 'abc123')
plt.show() # optional and Used to display the plot if running in a non-Jupyter environment
```
