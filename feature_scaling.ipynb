# Standardization

## StandardScaling
    - Scaler is used tp bring the data of a feature/columns in certain Range
    - xi= (xi-mean)/ standard deviation
    - does mean centering
    - mean of new data will alwaye be 0 and std. deviation is 1

```python
# split the data into train test
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test= train_test_split(x, df['scholarship'], test_size= 0.2)
```

```python
x_train
```

```python
y_train
```

```python
# scale the data

from sklearn.preprocessing import StandardScaler

scaler= StandardScaler()

scaler.fit(x_train)   #Always fit the data using x_train to know mean and standard deviation


x_train_scaled= scaler.transform(x_train)  #only scale the independent variable 
x_test_scaled= scaler.transform(x_test)
```

```python
x_train_scaled
```

```python
# converting the np array into a dataframe

x_scaled= pd.DataFrame(x_train_scaled, columns= x.columns)
x_scaled
```

```python
x.columns  # we can give columns name while creating dataframe by using the list
```

```python
#Train the model
from sklearn.linear_model import LogisticRegression

model= LogisticRegression()

model.fit(x_scaled, y_train)
```

```python
x_test_scaled= pd.DataFrame(x_test_scaled, columns= x.columns
                            )
x_test_scaled
```

```python
y_pred= model.predict(x_test_scaled)
y_pred
```

# Normalization
## MinMaxScaling
 - squeezes all data in range 0 to 1

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test= train_test_split(x, df['scholarship'], train_size= 0.8)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler= MinMaxScaler()

scaler.fit(x_train)

x_train_scaled= scaler.transform(x_train)
x_test_scaled= scaler.transform(x_test)
```

```python
data= pd.DataFrame(x_train_scaled, columns= x.columns)
data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x= data['cgpa'], y= data['iq'])
plt.title("After normalization")
plt.xlabel("CGPA")
plt.show()
```
