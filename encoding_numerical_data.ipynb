# Bining/ Discretization
    - Consist of 3 types: Uniform(Bins having same width), quantile(bins having same frequency) and kmeans bining

```python
x= df.iloc[:, [1, 2]]
y= df['Survived']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.2)

```

```python
from sklearn.preprocessing import KBinsDiscretizer 
from sklearn.compose import ColumnTransformer  #for binning

clf= ColumnTransformer(
    transformers=[
        ('clf1', KBinsDiscretizer(n_bins= 10, encode= 'ordinal', strategy= 'uniform'), ['Age']) ,  #n_bins= no. of bins/interval, encde= ordinal or one hot and strategy= uniform, discrete or kmeans
        ('clf2', KBinsDiscretizer(n_bins= 15, encode= "ordinal", strategy='quantile'), ['Fare'])
    ], remainder= 'passthrough', verbose_feature_names_out= False
)
```

```python
clf.fit(x_train)
```

```python
x_train_clf= clf.transform(x_train)
x_test_clf= clf.transform(x_test)
```

```python
clf.set_output(transform= "pandas")
```

```python
x_train_clf   #here each numerical value is replaced by its  inverval and ordinal encpding is done in that onterval
```

```python
clf.get_feature_names_out()
```

# Binarization
- Used in special cases only
- It converts a column into binary value depending on the threshold
- it copy = true then it will create new column. If copy is false then it will make changes in same column

```python
from sklearn.preprocessing import Binarizer
 
df= pd.read_csv("./datasets/train.csv")
```

```python
df.head()
```

```python
df.dropna(inplace= True)
```

```python
df= df.iloc[:, [ 1, 6, 7]]
```

```python
df.head()
```

```python
df['Siblings']= df['SibSp'] + df['Parch']
```

```python
df.head()
```

```python
df.drop(columns= ['SibSp', 'Parch'], inplace= True)
```

```python
df.head()
```

```python
# Appllying on Siblings if 0 then 0 ie NO siblings, If more than 0 than siblings present= 1
bnz= Binarizer(threshold= 0, copy= False)
```

```python
df_new= bnz.fit_transform(df)
```

```python
pd.DataFrame(df_new, columns= bnz.get_feature_names_out()).head()
```
