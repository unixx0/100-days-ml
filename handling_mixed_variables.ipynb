```python
## for the column not having the numeric value and Alphabetic value in same cell

#extract numeric part
df['Mixed_Num_Letter_Numeric']= pd.to_numeric(df['Mixed_Num_Letter'], errors= "coerce", downcast= 'integer')
```

```python
# extract the categorical part
df['Mixed_Num_Letter_category']= np.where(df['Mixed_Num_Letter_Numeric'].isnull(), df['Mixed_Num_Letter'], np.nan)
```

```python
df.head()
```

```python
# for having numerical and categorical in same cell
df['Code_Info_Numeric']= df['Code_Info'].str.extract('(\d+)')   #captures numeric part
df["Code_Info_Categorical"]= df['Code_Info'].str[0]    #extracts first element
```
