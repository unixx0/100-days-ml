# Encoding Categorical Data

- The process of changing categorical columns into numerical columns is called Encoding Categorical Data

```python
import pandas as pd
```

```python
df= pd.read_csv("./datasets/students.csv")
```

```python
df.head()
```

```python
df= df.iloc[:, 2:]
df.sample(5)
```

## Ordinary Enconding
- Ordinary encoding is done for the column having  ordinary categorical data i.e the category in column are in order
- Education_level and Satisfaction are columns containing ordinary categorical data

```python
# Splitting the data into train test split

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test= train_test_split(df.iloc[:, [2, 3]], df['Purchase'], train_size= 0.7)
```

```python
x_test
```

```python
# ordinary encoding
from sklearn.preprocessing import OrdinalEncoder

oe= OrdinalEncoder(categories= [['High School', 'Bachelor', 'Master', 'PhD'], ['Poor', "Average", "Good", "Excellent"]])  # no of list = no. of ordinal categorical columns. ['High School', 'Bachelor', 'Master', 'PhD']= [0,1,2,3]

# Always fit using train data
oe.fit(x_train)


#transform both train and test data
x_test= oe.transform(x_test)
x_train= oe.transform(x_train)
```

```python
pd.DataFrame(x_test, columns= df.iloc[:, [2, 3]].columns )
```

## Label Encoding

- Label Encoding is same like ordinal Encoding ; ordinal encoding is for input columns whereas label encoding is for output columns.

- Note: We can use Ordinal/ Label encoding in columns having Nominal catagories if the no of category is exactly 2(binary) like yes/no etc.

```python
from sklearn.preprocessing import LabelEncoder

le= LabelEncoder()  # we donot provide any arguments(order) here like in Ordinal encoder. It automatically assigns value to the catagories
le.fit(y_train)

y_test= le.transform(y_test)
y_train= le.transform(y_train)
```

```python
y_train
```

# Column Transformer
- Transforming each column on the basis of its property and finally merging it is so inefficient. Thats why we use column transformer whichh transforms all the columns with one pipeline the way we want so we donot have to concat each one

```python
df= pd.read_csv("./datasets/customer.csv")
```

```python
df.sample(5)
```

```python
#df= df.drop(columns= "CustomerID")
```

```python
df.sample(5)
```

```python
x_train, x_test, y_train, y_test= train_test_split(df.iloc[:, :-1], df['Purchased'], train_size= 0.8)
```

```python
x_train.head()
```

```python
x_train['Education'].value_counts()
```

```python
y_train.head()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, MinMaxScaler
transformer= ColumnTransformer(
    transformers=[
('tnf1', MinMaxScaler(), ['Age', 'Spending_Score']),
('tnf2', StandardScaler(), ['Annual_Income']),
('tnf3', OneHotEncoder(sparse_output= False, drop= 'first', dtype= np.int32), ['Gender', 'City']),
('tnf4', OrdinalEncoder(categories= [['High School', "Bachelor's", "Master's", "PhD"], ['Low', "Medium", "High", "Very High"]]), ['Education','Satisfaction_Level'])
    ], remainder= "passthrough"
)


#remainder helps us to know what to do with the column we ddint transfoorm. If it is passthrough then it wont do anything to column and if it is from thwn it will drop the column
#transformer= [(name, obect, list of colm name where transformer should be appied)]
```

```python
x_train= transformer.fit_transform(x_train)
```

```python
transformer.get_feature_names_out()
```
